```python
from collections import Counter
```

```python
L = ['Mark', 'Johnny', 'David', 'Mark', 'Luke', 'Johnny', 'Mark']
```

```python
c = Counter(L)
```

```python
c
```

```python
c.keys()
```

```python
c.values()
```

```python
c.get('Mark')
```

```python
for i in c.elements():
    print(i)
```

```python
c.update({'John':4})
```

```python
for i in c.elements():
    print(i)
```

```python
c.pop('John')
```

```python
c.most_common(1)  #the top most common item
```

```python
c.most_common(2)  #two most common
```
